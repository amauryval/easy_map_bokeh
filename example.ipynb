# Bokeh for Map

## Prepare input data from geojson

```python
import geopandas as gpd


multipolygons = gpd.GeoDataFrame.from_file("tests/fixtures/multipolygons.geojson")
polygons = gpd.GeoDataFrame.from_file("tests/fixtures/polygons.geojson")
points = gpd.GeoDataFrame.from_file("tests/fixtures/points.geojson")
linestrings = gpd.GeoDataFrame.from_file("tests/fixtures/linestrings.geojson")
multilinestrings = gpd.GeoDataFrame.from_file("tests/fixtures/multilinestrings.geojson")
```

## Map them !

```python
from bokeh.plotting import output_notebook
from bokeh.plotting import show

from bfm import BokehForMap

output_notebook()


my_map = BokehForMap("My beautiful map", width=640, height=800)

my_map.add_polygons(
    polygons,
    fill_color="red",
    legend="Polygons"
)

my_map.add_polygons(
    multipolygons,
    fill_color="orange",
    legend="MultiPolygons"
)

my_map.add_lines(
    linestrings,
    color="blue",
    legend="LineStrings"
)

my_map.add_lines(
    multilinestrings,
    color="green",
    legend="MultiLineStrings"
)

my_map.add_points(
    points,
    fill_color="brown",
    legend="points"
)

show(my_map.figure)
```
