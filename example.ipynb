# Bokeh for Map

## Prepare input data from geojson and map them

```python
import geopandas as gpd


from bokeh.plotting import output_notebook
from bokeh.plotting import show

from bokeh_for_map import BokehForMap

output_notebook()


my_map = BokehForMap("My beautiful map", width=640, height=800)

# POLYGONS
polygons = gpd.GeoDataFrame.from_file("tests/fixtures/polygons.geojson")
bokeh_polygons = my_map.format_features(polygons)
my_map.add_polygons(
    bokeh_polygons["data"],
    fill_color="red",
    legend="Polygons"
)

# MULTIPOLYGONS
multipolygons = gpd.GeoDataFrame.from_file("tests/fixtures/multipolygons.geojson")
bokeh_multipolygons = my_map.format_features(multipolygons)
my_map.add_polygons(
    bokeh_multipolygons["data"],
    fill_color="orange",
    legend="MultiPolygons"
)

# LINESTRINGS
linestrings = gpd.GeoDataFrame.from_file("tests/fixtures/linestrings.geojson")
bokeh_linestrings = my_map.format_features(linestrings)
my_map.add_lines(
    bokeh_linestrings["data"],
    color="blue",
    legend="LineStrings"
)

# MULTILINESTRINGS
multilinestrings = gpd.GeoDataFrame.from_file("tests/fixtures/multilinestrings.geojson")
bokeh_multilinestrings = my_map.format_features(multilinestrings)
my_map.add_lines(
    bokeh_multilinestrings["data"],
    color="green",
    legend="MultiLineStrings"
)

# POINTS
points = gpd.GeoDataFrame.from_file("tests/fixtures/points.geojson")
bokeh_points = my_map.format_features(points)
my_map.add_points(
    bokeh_points["data"],
    fill_color="brown",
    legend="points"
)

show(my_map.figure)
```
